```python
from math import factorial
def solution(balls, share):
    answer = factorial(balls) // (factorial(balls - share) * factorial(share))
    return answer
```

```python
balls = 3
share = 2
solution(balls, share)
```

```python
balls = 5
share = 3
solution(balls, share)
```
